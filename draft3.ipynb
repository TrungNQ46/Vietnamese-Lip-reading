```python
import os

output_folder_video = 'data_output/video_non_audio'
# pl_folders = ['DONGC2', 'BAIGI1', 'BAIGI2', 'BAIGI3']

pl_folders = [f.name for f in os.scandir(output_folder_video) if f.is_dir()]
print(pl_folders)
for pl_folder in pl_folders:
    if not os.path.exists(output_folder_video + '/' + pl_folder):
        print(f"Error: The folder {input_folder} does not exist.")
    else:
        input_folders = [f.name for f in os.scandir(output_folder_video + '/' + pl_folder) if f.is_dir()]
        for input_folder in input_folders:     
            path = output_folder_video + '/' + pl_folder + '/' + input_folder  


            # Get all files in the folder
            all_files = os.listdir(path)

            # Create sets of .mp4 and .csv filenames without extensions
            mp4_files = {f[:-11] for f in all_files if f.endswith('.mp4')}
            csv_files = {f[:-4] for f in all_files if f.endswith('.csv')}

            # Identify .mp4 files that don't have corresponding .csv files
            mp4_to_delete = mp4_files - csv_files
            # print(mp4_to_delete)
            # Delete the unmatched .mp4 files
            for file in mp4_to_delete:
                mp4_file_path = os.path.join(path, file + '_output' + '.mp4')
                if os.path.exists(mp4_file_path):
                    # os.remove(mp4_file_path)
                    print(f"Deleted: {mp4_file_path}")
```

```python
import os
import cv2
input_folder_video = 'data_output/video'
output_folder_video = 'data_output/Vlips_v1_video_non_audio'
# pl_folders = ['DONGC2', 'BAIGI1', 'BAIGI2', 'BAIGI3']

pl_folders = [f.name for f in os.scandir(output_folder_video) if f.is_dir()]
print(pl_folders)
for pl_folder in pl_folders:
    if not os.path.exists(output_folder_video + '/' + pl_folder):
        print(f"Error: The folder {input_folder} does not exist.")
    else:
        input_folders = [f.name for f in os.scandir(output_folder_video + '/' + pl_folder) if f.is_dir()]
        for input_folder in input_folders:     
            path = output_folder_video + '/' + pl_folder + '/' + input_folder  


            # Get all files in the folder
            all_files = os.listdir(path)

            # Create sets of .mp4 and .csv filenames without extensions
            mp4_files = {f[:-4] for f in all_files if f.endswith('.mp4')}
            csv_files = {f[:-4] for f in all_files if f.endswith('.csv')}

            # Identify .mp4 files that don't have corresponding .csv files
            # mp4_to_delete = mp4_files - csv_files
            # print(mp4_to_delete)
            # Delete the unmatched .mp4 files
            for file in mp4_files:
                mp4_file_path = os.path.join(path, file + '.mp4')
                if os.path.exists(mp4_file_path):
                    # os.remove(mp4_file_path)
                    video = cv2.VideoCapture(mp4_file_path)
                    total_frames_1 = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
                    video.release()
                    video2 = cv2.VideoCapture(os.path.join(input_folder_video + '/' + pl_folder + '/' + input_folder, file + '.mp4'))
                    total_frames_2 = int(video2.get(cv2.CAP_PROP_FRAME_COUNT))
                    video2.release()
                    if total_frames_1 - total_frames_2 != 0:

                        print(f"Deleted: {mp4_file_path}")
```
